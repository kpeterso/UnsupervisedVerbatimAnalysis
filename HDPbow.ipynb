```python
import re
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
#from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from gensim import corpora, models
import numpy as np
```

```python
f=open("Data/comments.txt",'r')
docSet = f.readlines()
f.close()
stops = set(stopwords.words("english"))
```

```python
# Create p_stemmer of class PorterStemmer
#p_stemmer = PorterStemmer()
lemmatizer=WordNetLemmatizer()
#create a tokenizer
tokenizer = RegexpTokenizer(r'\w+')
#create a list of stop words
stops = set(stopwords.words("english"))
additional_stopwords=[]
additional_stopwords.append("car")
additional_stopwords.append("nissan")
additional_stopwords.append("altima")
additional_stopwords.append("sentra")
additional_stopwords.append("vehicle")
additional_stopwords.append("drive")
all_stops=stops.union(set(additional_stopwords))


texts = []
for i in docSet:
    # clean and tokenize document string
    raw = i.lower()
    #remove non-letter chars
    letters_only = re.sub("[^a-zA-Z]", " ", raw)
    #tokenize words
    tokens = tokenizer.tokenize(letters_only)

    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in all_stops]
    
    # stem tokens
    #stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    #lemmatize tokens
    lemmad_tokens = [lemmatizer.lemmatize(i) for i in stopped_tokens]
    
    # add tokens to list
    #texts.append(stemmed_tokens)
    texts.append(lemmad_tokens)
```

```python
# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)
dictionary.filter_extremes(no_below=5, no_above=.5, keep_n=None)
dictionary.compactify()

# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
#adjust BOW with tf-idf weighting
#tfModel=models.TfidfModel(corpus)
#corpus_tfidf = tfModel[corpus]
```

```python
#HDP model, tries to find new topics
hdpmodel = models.hdpmodel.HdpModel(corpus, dictionary)
```

```python
hdpmodel.print_topics(topn=5)
```
